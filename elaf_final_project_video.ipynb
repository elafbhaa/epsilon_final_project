# Video Game Sales and Ratings Data

##### This dataset provides comprehensive data on video game sales, ratings, and user insights globally, including details such as the game's name, platform, year of release, genre, publisher, sales in different regions, critic and user scores, and ESRB rating.




###  Column Descriptions

| Column | Description |
|--------|------------|
| Index | Unique identifier for each video game record. |
| Name | The name of the video game.|
| Platform | The platform on which the game is available, such as PC, PS4, Xbox, etc. |
| Year_of_Release | The year in which the game was released. |
| Genre | The genre of the game, such as Action, Sports, etc.|
| Publisher|  The company that published the game.  |
| NA_Sales| The sales of the game in North America, in millions. |
| EU_Sales| The sales of the game in Europe, in millions.|
| JP_Sales| The sales of the game in Japan, in millions. |
| Other_Sales| The sales of the game in other regions, in millions.|
| Global_Sales| The total global sales of the game, in millions. |
| Critic_Score| The average score given to the game by critics.|
| Critic_Count| The number of critics who reviewed the game.|
| User_Score| The number of critics who reviewed the game.|
| User_Count | The average score given to the game by users.|
| Developer| The company that developed the game.|
| Rating| The ESRB rating of the game, such as E for Everyone, T for Teen, or M for Mature.|

#### Additional Features to Enhance Understanding and better explanation
###### Sales Category: Define a function to categorize sales and apply it in the existing global_sales to create a new feature named sales_category.
###### Game Age Category: Calulate game age and define a function to categorize game age is it new, recent, old or classic.



```python
# Sales Category

df['global_sales'] = pd.to_numeric(df['global_sales'], errors='coerce')   # to ensure global_sales is numeric

def categorize_sales(value):
    
    if 0 <= value < 25:
        return 'low'
    elif 25 <= value < 50:
        return 'medium'
    elif 50 <= value < 75:
        return 'high'
    elif 75 <= value <= 100:
        return 'best_seller'
    else:
        return 'out_of_range'  # optional for values > 100

# Apply the function to create a new column
df['sales_category'] = df['global_sales'].apply(categorize_sales)

print(df[['name','global_sales','sales_category']].head())
```

```python
# Game Age Category

current_year= datetime.datetime.now().year
df['game_age']=current_year - df['year_of_release']      # Calulate Gmae age

print(df[['year_of_release','game_age']].head())
```

```python
def categorize_age(value):
    
    if value < 5:
        return 'new'
    elif 5 <= value < 15:
        return 'recent'
    else:
        return 'old'
    
# Apply the function to create a new column
df['game_age_category'] = df['game_age'].apply(categorize_age)

print(df[['game_age_category','game_age']].head())
```

### Machine Learning

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_validate, ShuffleSplit
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import RobustScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from category_encoders import BinaryEncoder
from sklearn.compose import ColumnTransformer
import joblib
```

```python
df = pd.read_csv("clean_video_games.csv")

# Ensure target is numeric and aligned
df["global_sales"] = pd.to_numeric(df["global_sales"], errors="coerce")
df = df.dropna(subset=["global_sales"])

X = df.drop(columns=["global_sales", "sales_category", "game_age", "game_age_category"], errors="ignore")
y = df["global_sales"].astype(float)

# Optional: filter out zero-sales rows to avoid NaN R²
df = df[df["global_sales"] > 0]
X = df.drop(columns=["global_sales"], errors="ignore")
y = df["global_sales"].astype(float)
```

```python
# 3. Preprocessing Pipeline
# ============================================
num = ['year_of_release','na_sales','eu_sales','jp_sales','other_sales','user_score']
cat_ohe = ['genre','rating']
cat_be = ['name','platform','publisher','developer']

num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", RobustScaler())
])


ohe_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("ohe", OneHotEncoder(drop="first", sparse_output=False, handle_unknown="ignore"))
])



be_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="constant", fill_value="Unknown")),
    ("be", BinaryEncoder())
])

preprocessing_pipe = ColumnTransformer([
    ("num", num_pipeline, num),
    ("ohe", ohe_pipeline, cat_ohe),
    ("be", be_pipeline, cat_be)
], remainder="drop")
```

```python
model_zoo = [
    ("Linear Regression", LinearRegression()),
    ("KNN", KNeighborsRegressor(n_neighbors=10)),
    ("SVR", SVR(kernel="rbf")),
    ("Decision Tree", DecisionTreeRegressor(max_depth=7, random_state=42)),
    ("Random Forest", RandomForestRegressor(random_state=42))
]
```

```python
cv_split = ShuffleSplit(n_splits=5, test_size=0.2, random_state=42)

results = []
best_model = None
best_score = -np.inf

for name, model in model_zoo:
    model_pipeline = Pipeline([
        ("Preprocessing", preprocessing_pipe),
        ("Model", model)
    ])
    
    cv = cross_validate(
        model_pipeline,
        X,
        y,
        cv=cv_split,
        scoring=["r2", "neg_mean_absolute_error", "neg_root_mean_squared_error"],
        return_train_score=True,
        n_jobs=-1,
        error_score="raise"
    )
    
    results.append({
        "Model": name,
        "Train R²": np.nanmean(cv["train_r2"]),
        "Test R²": np.nanmean(cv["test_r2"]),
        "Test MAE": -np.nanmean(cv["test_neg_mean_absolute_error"]),
        "Test RMSE": -np.nanmean(cv["test_neg_root_mean_squared_error"])
    })
    
    if np.nanmean(cv["test_r2"]) > best_score:
        best_score = np.nanmean(cv["test_r2"])
        best_model = model_pipeline

results_df = pd.DataFrame(results)
display(results_df)
```

```python
best_model.fit(X, y)
joblib.dump(best_model, "video_game_model.pkl")
print("Best model saved as video_game_model.pkl")
```
